### Определим для какого вида объекта общественного питания характерно сетевое распространение?

```python
# построение графика
displot = sns.displot(x = "object_type", hue='chain', data = data,  multiple='dodge', palette='hls')
plt.gcf().set_size_inches(18, 6)

plt.title('Сетевое распространение по видам объектов')
displot.ax.tick_params(axis='x', labelrotation = 90)
plt.show()
```

График показывает:
* сетевое распространение более характерно для кафе, предприятий быстрого обслуживания и ресторанов; 
* среди столовых практически нет сетей. 

Построим сводную таблицу и рассчитаем долю сетевых заведений от общего количества заведий по видам. 

```python
# создание сводной таблицы
chain_data = data.pivot_table(
    index='object_type',  
    values=  'chain',
    aggfunc='mean'  
).reset_index().rename(columns={False: 'not_chain', True:'chain'})

chain_data['chain'] = (chain_data['chain']*100).round(2)    #.apply(lambda x: format(x, '.2%'))
chain_data = chain_data.sort_values(by='chain', ascending=False) # вывод на экран таблицы chain_data
chain_data
```

```python
# построение графика на основании таблицы chain_data
fig = px.bar(chain_data, x='object_type', y='chain', color='object_type', text='chain')
fig.update_layout(
    title='Доля сетевых заведений от общего числа, по видам объектов',
    xaxis_title="Тип объекта",
    yaxis_title="Доля")
fig.show()
```

* Наибольшая доля сетевых точек у предприятий быстрого обслуживания - 41,52%. 
* От 20 до 30% сетевых заведений среди магазинов (отдел кулинарии), ресторанов и кафе. 
* Для баров, буфетов и столовых не характерно сетевое распространение.

### Что характерно для сетевых заведений: много заведений с небольшим числом посадочных мест в каждом или мало заведений с большим количеством посадочных мест?

Поместим в переменную `chain_rest` информацию исключительно о сетевых точках общественного питания.

```python
# создание среза и группировка по наименованию
chain_rest = data.query('chain==True').groupby('object_name').agg({'object_name':'count', 'number':'median'})
```

```python
chain_rest['object_name'].describe() # получение списка характерных значений
```

```python
chain_rest['number'].describe() # получение списка характерных значений
```

```python
# построение диаграммы рассеяния
scatterplot = sns.scatterplot(data=chain_rest, x="number", y="object_name")
plt.gcf().set_size_inches(12, 6)
scatterplot.axes.set_title("Совместное распреление количества посадочных мест и количества заведений сети", fontsize=16)
scatterplot.set_xlabel("Медианное количество посадочных мест", fontsize=12)
scatterplot.set_ylabel("Кол-во заведений одной сети", fontsize=12)

plt.show()
```

График показал:
* для сетевых заведений более характерно открывать малое количество заведений с небольшим количеством посадочных мест; 
* в среднем сети имеют по 4 заведения; 
* среднее количество посадочных мест в одном заведении - 59.

```python
chain_rest.query('object_name > 100') # построение среза по количеству заведений
```

 Но есть так же, сети, в которых открыто более 100 зеведений (KFC, Бургер Кинг, Макдоналдс, Шоколадница). 

```python
chain_rest.query('object_name < 2') # построение среза по количеству заведений
```

В таблице есть 450 заведений, которые отнносятся к сетевым, но имеют всего одно заведение в сети. Построим диаграмму размаха без учета этих заведений.

```python
chain_rest_query = chain_rest.query('object_name >= 2') # срез по количеству заведений в сети 
```

```python
# построение диаграммы размаха
scatterplot = sns.scatterplot(data=chain_rest_query, x="number", y="object_name")
plt.gcf().set_size_inches(12, 6)
scatterplot.axes.set_title("Совместное распреление количества посадочных мест и количества заведений сети", fontsize=16)
scatterplot.set_xlabel("Медианное количество посадочных мест", fontsize=12)
scatterplot.set_ylabel("Количество заведений одной сети", fontsize=12)

plt.show()
```

```python
chain_rest_query['object_name'].describe() # получение списка характерных значений
```

```python
chain_rest_query['number'].describe() # получение списка характерных значений
```

После очистки данных от выбивающихся значений, график практически не изменился. В среднем сетевые заведения имеют по 9 точек обшественного питания, в одной точке в среднем 54 места.

### Для каждого вида объекта общественного питания опишем среднее количество посадочных мест. Определим какой вид предоставляет в среднем самое большое количество посадочных мест? Построим график.

 С помощью диаграммы размаха посмотрим распределение посадочных мест по объектам общественного питания.

```python
# построение диаграммы размаха
fig = px.box(data, x="number", y="object_type", title="", color = 'object_type' )
fig.update_layout(
    title='Распределение количества посадочных мест по видам объектов',
    yaxis_title="Тип объекта",
    xaxis_title="Количество")
fig.show()
```

Графики получились очень сжатыми из за большого количества выбросов. По некоторым видам заведений выбросы достигают 1000 и более посадочных мест. Найдем среднее количество посадочных мест для каждого вида объекта общественного питания и отразим их с помощью столбчатой диаграммы.

```python
# построение столбчатой диаграммы
type_rest = data.groupby('object_type').agg({'number':'mean'}).reset_index().sort_values(by='number')

fig = px.bar(type_rest, x='object_type', y='number', color='object_type') #, text='number')
fig.update_layout(
    title='Среднее количество посадочных мест по видам объектов',
    xaxis_title="Тип объекта",
    yaxis_title="Количество")
fig.show()
```

* Наибольшее количество посадочных мест у столовых (130,34), ресторанов (96,88) и буфетов (51,51).
* Меньше всего - в отделах кулинарии в магазинах (5,59), в закусочных (7,83) и кафетериях (9,18).

### Выделим в отдельный столбец информацию об улице из столбца `address`

```python
data['street'] = data['address'].str.split(",", expand=True)[1] # выделение улицы в отдельный столбец
```

Не во всех случаях улицы выделились корректно. Скорее всего это произошло с адресами, записанными не стандартным образом. Заменим пропуском значения в стобце `street`, в которых вместо названия улицы встречается город/поселение/деревня/территирия/дом. Методом `str.strip()` удалим лишние пробелы, образовавшиеся после выделения улиц.

```python
def correct_street(street):  # функция для замены пропуском некорректных названий улиц
    if 'город' in street:
        return 'NaN'
    if 'оселение' in street:
        return 'NaN'
    if 'еревня'in street:
        return 'NaN'
    if 'Город' in street:
        return 'NaN'
    if 'ерритория' in street:
        return 'NaN'
    if 'дом' in street:
        return 'NaN'
    else:
        return street
```

```python
data['street'] = data['street'].apply(correct_street) # применение функции correct_street к столбцу street
data['street'] = data['street'].str.strip() # удаление пробелов в столбце street
```

### Найдем число улиц с одним объектом общественного питания. Воспользуемся внешней информацией и ответим на вопрос — в каких районах Москвы находятся эти улицы?

```python
# создание сводной таблицы
one_object_name = data_query.pivot_table(index='street', values='object_name', aggfunc='count').reset_index().sort_values(by='object_name')
one_object_name.query('object_name == 1').count()
```

На 542 улицах расположен всего один объект общественного питания.

```python
one_object_name = one_object_name.query('object_name == 1') # создание таблицы с условием 
one_object_name.head() # просмотр первых строк таблицы
```

Для того, чтобы узнать в каких районах Москвы расположены данные улицы, присоединим к таблице `one_object_name` таблицу с районами Москвы `df2`. Результат сохраним в переменной `one_object_name_with_district`.

```python
one_object_name_with_district.head() # просмотр первых строк таблицы
```

Сгруппируем полученную таблицу по районам, и посмотрим в каких районах Москвы больше всего улиц с единственным заведением общественного питания.
